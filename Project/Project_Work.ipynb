```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import nengo
import nengo_dl

training_time = False

# Import Dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize Dataset Pixel Values [0,1]
train_images, test_images = train_images / 255.0, test_images / 255.0

# For Reference
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Visualize Sample Images and Classifications
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]])
plt.savefig('sample_CIFAR.svg')
plt.show()

# Create a basic convolutional network for CIFAR-10 Classification
mInput = keras.Input(shape=(32, 32, 3))
mConv1 = keras.layers.Conv2D(32, (3, 3), activation=tf.nn.relu, input_shape=(32, 32, 3))(mInput)
mPool1 = keras.layers.MaxPooling2D((2, 2))(mConv1)
mConv2 = keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu)(mPool1)
mPool2 = keras.layers.MaxPooling2D((2, 2))(mConv2)
mConv3 = keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu)(mPool2)
mFlat = keras.layers.Flatten()(mConv3)
mDense1 = keras.layers.Dense(64, activation=tf.nn.relu)(mFlat)
mDense2 = keras.layers.Dense(10)(mDense1)
model = keras.Model(inputs=mInput, outputs=mDense2)
model.compile(optimizer='adam', loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['sparse_categorical_accuracy'])

# Training for Nengo-DL Realized Models is Equivalent to Keras Training, Run Keras Training or Evaluation
if training_time:
    model.fit(train_images, train_labels, epochs=10,
              validation_data=(test_images, test_labels))
    model.save_weights("/home/j2chouin/Dev/SYDE552/Project/trained_cnn.weights.h5")
else:
    model.load_weights("/home/j2chouin/Dev/SYDE552/Project/trained_cnn.weights.h5")
    print("KERAS MODEL BASELINE PERFORMANCE:")
    model.evaluate(test_images, test_labels)

# Outer-Layer Simulation Parameters
neuron_set = [nengo.SpikingRectifiedLinear(), nengo.LIF(), nengo.AdaptiveLIF()]
synapse_set = [0.005, 0.010, 0.015, 0.020]
performance_map = np.zeros((len(neuron_set),len(synapse_set),2))

f, ((pl1, pl2, pl3), (pl4, pl5, pl6), (pl7, pl8, pl9), (pl10, pl11, pl12)) = plt.subplots(4,3)
pls = [pl1, pl4, pl7, pl10, pl2, pl5, pl8, pl11, pl3, pl6, pl9, pl12]
f.subplots_adjust(right=3.0, top=4.0)
for n_idx in range(len(neuron_set)):
    for s_idx in range(len(synapse_set)):
        # Nengo Hyper-Parameters
        num_megabatch, megabatch_size, minibatch_size, n_steps, synapse, neuron = 4, 250, 25, 100, synapse_set[s_idx], neuron_set[n_idx]
        
        # Simulation Config Output
        print(f"RUNNING SIMULATION FOR NEURON: {neuron}, SYNAPSE: {synapse}")
        
        # Generate a Nengo-DL Equivalent Representation that Utilizes Varying Spiking Neuron Types
        cnv = nengo_dl.Converter(model, max_to_avg_pool=True, scale_firing_rates=415, swap_activations={tf.nn.relu: neuron}, synapse=synapse)

        # Nengo-DL Simulation with MiniBatch Size
        with nengo_dl.Simulator(cnv.net, minibatch_size=minibatch_size) as sim:
            # Flatten Data
            f_train_images = train_images.reshape((train_images.shape[0], 1, -1))
            f_train_labels = train_labels.reshape((train_labels.shape[0], 1, -1))
            f_test_images = test_images.reshape((test_images.shape[0], 1, -1))
            f_test_labels = test_labels.reshape((test_labels.shape[0], 1, -1))

            # Simulation Compilation Parameters and Metrics
            sim.compile(optimizer='adam', loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['sparse_categorical_accuracy'])
            
            # Evaluate 50ms Exposure to Input Images (250 Images at a Time -- 1000 Total)
            scaccy = []
            scloss = []
            for i in range(num_megabatch):
                idx_start = i * megabatch_size
                idx_end = (i + 1) * megabatch_size
                results = sim.evaluate({cnv.inputs[mInput]: np.squeeze(np.tile(f_test_images[idx_start:idx_end,None,:],(1,n_steps,1)))}, {cnv.outputs[mDense2]: np.tile(f_test_labels[idx_start:idx_end,None,:],(1,n_steps,1)).reshape(-1,n_steps,1)})
                scaccy.append(results['probe_sparse_categorical_accuracy'])
                scloss.append(results['probe_loss'])
            avg_accuracy = np.mean(scaccy)
            avg_loss = np.mean(scloss)
            print(f"NENGO PERFORMANCE: {avg_accuracy}, {avg_loss}")
            performance_map[n_idx,s_idx,0] = avg_accuracy
            performance_map[n_idx,s_idx,1] = avg_loss
        sim.close()
        with nengo_dl.Simulator(cnv.net) as sim:
            data = sim.predict({cnv.inputs[mInput]: np.tile(f_test_images[0,None,:],(1,100,1))})
            pls[n_idx * 4 + s_idx].plot(np.squeeze(data[cnv.outputs[mDense2]]))
            pls[n_idx * 4 + s_idx].set(title=f"{neuron}, {synapse}")
            pls[n_idx * 4 + s_idx].grid()
        sim.close()
```
